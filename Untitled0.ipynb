```python
import geopandas as gdp
import pandas as pd
import matplotlib.pyplot as plt
from pyproj import Proj
import contextily as ctx
```

```python
municipalities = gdp.read_file("KOMMUNE.shp")
municipalities.head()
```

```python
bydel = gdp.read_file("bydel.shp")
bydel.head()
```

```python
bydel.plot()
```

```python
municipalities.plot()
```

```python
municipalities.boundary.plot()
```

```python
mun = municipalities.dissolve(by='KOMNAVN')
```

```python
mun.info(verbose=False)
```

```python
mun_data=pd.read_excel("https://github.com/jonasgr5/FSA_geodata/raw/main/spatialdata_baba_omr_ny.xlsx")
mun_data.head()
```

```python
mun_fun = mun.merge(mun_data, left_on='KOMNAVN', right_on="omr", how='left', validate="one_to_one")
mun_fun.head()
```

```python
from google.colab import files
municipalities.to_csv('output2.csv', encoding = 'utf-8-sig') 
files.download('output2.csv')
```

```python
mun_fun.plot(column='Baba_Bydelsmor',edgecolor='black', figsize=(14,14))
from google.colab import files
plt.savefig("bydel2.pdf")
files.download("bydel2.pdf")
```
